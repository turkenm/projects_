## Task - 2

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df_21 = pd.read_csv("E0_2021.csv")
df_20 = pd.read_csv("E0_1920.csv")
df_19 = pd.read_csv("E0_1819.csv")
pd.set_option("display.max_columns", 101)
```

```python
odds = ["B365H","B365D","B365A","BWH","BWD","BWA","PSH","PSD","PSA","VCH","VCD","VCA"]
```

```python
odds_df = pd.concat([df_21[odds], df_20[odds], df_19[odds]], ignore_index = True)
full_df = pd.concat([df_21, df_20, df_19], ignore_index = True)
```

```python
full_df
```

```python
probs_df = 1/odds_df
```

```python
b365df = probs_df[["B365H","B365D","B365A"]]
bwdf = probs_df[["BWH","BWD","BWA"]]
psdf = probs_df[["PSH","PSD","PSA"]]
vcdf = probs_df[["VCH","VCD","VCA"]]
```

Normalizing odds.

```python
a = b365df.div(b365df.sum(axis=1), axis=0)
b = bwdf.div(bwdf.sum(axis=1), axis=0)
c = psdf.div(psdf.sum(axis=1), axis=0)
d = vcdf.div(vcdf.sum(axis=1), axis=0)
```

```python
norm_prob_df = pd.concat([a,b,c,d], axis = 1)
```

```python
norm_prob_df.head()
```

```python
norm_prob_df["B365dif"] = norm_prob_df["B365H"] - norm_prob_df["B365A"]
norm_prob_df["BWdif"] = norm_prob_df["BWH"] - norm_prob_df["BWA"]
norm_prob_df["PSdif"] = norm_prob_df["PSH"] - norm_prob_df["PSA"]
norm_prob_df["VCdif"] = norm_prob_df["VCH"] - norm_prob_df["VCA"]
```

Converting categorical results to binary values for "Draw". If match result "Draw" = 1

```python
full_df["FTR"].replace({"H": "0", "A": "0", "D" : "1"}, inplace=True)
full_df = full_df.astype({'FTR': 'float32'})
norm_prob_df["FTR"] = full_df["FTR"]
```

```python
norm_prob_df.head()
```

```python
k = np.linspace(-0.9,0.9, num=17)
```

```python
bins = np.linspace(-1, 1, num=20)
norm_prob_df["bins"] = pd.cut(norm_prob_df["BWdif"], bins, right=True, labels=False)
bin_df = norm_prob_df[["bins", "FTR"]].groupby(["bins"]).mean()

plt.scatter(norm_prob_df["BWdif"], norm_prob_df["BWD"], marker='.',label="BW",c="m")
plt.plot(k,bin_df["FTR"],color="y",label="Actual", marker = "o", linestyle = "dotted")
plt.xlabel("P(Home)-P(Away)")
plt.ylabel("P(Draw)")
plt.title("BetWin Figure 1")
plt.legend(loc="lower center")
plt.show()
```

```python
bins = np.linspace(-1, 1, num=20)
norm_prob_df["bins"] = pd.cut(norm_prob_df["B365dif"], bins, right=True, labels=False)
bin_df = norm_prob_df[["bins", "FTR"]].groupby(["bins"]).mean()

plt.scatter(norm_prob_df["B365dif"], norm_prob_df["B365D"], marker='.',label="Bet365",c="red")
plt.plot(k,bin_df["FTR"],color="y",label="Actual", marker = "o", linestyle = "dotted")
plt.xlabel("P(Home)-P(Away)")
plt.ylabel("P(Draw)")
plt.title("Bet365 Figure 2")
plt.legend(loc="lower center")
plt.show()
```

```python
bins = np.linspace(-1, 1, num=20)
norm_prob_df["bins"] = pd.cut(norm_prob_df["PSdif"], bins, right=True, labels=False)
bin_df = norm_prob_df[["bins", "FTR"]].groupby(["bins"]).mean()

plt.scatter(norm_prob_df["PSdif"], norm_prob_df["PSD"], marker='.',label="PS",c="b")
plt.plot(k,bin_df["FTR"],color="y",label="Actual", marker = "o", linestyle = "dotted")
plt.xlabel("P(Home)-P(Away)")
plt.ylabel("P(Draw)")
plt.title("Pinnacle Figure 3")
plt.legend(loc="lower center")
plt.show()
```

```python
bins = np.linspace(-1, 1, num=19)
norm_prob_df["bins"] = pd.cut(norm_prob_df["VCdif"], bins, right=True, labels=False)
bin_df = norm_prob_df[["bins", "FTR"]].groupby(["bins"]).mean()

plt.scatter(norm_prob_df["VCdif"], norm_prob_df["VCD"], marker='.',label="VC",c="r")
plt.plot(k,bin_df["FTR"],color="y",label="Actual", marker = "o", linestyle = "dotted")
plt.xlabel("P(Home)-P(Away)")
plt.ylabel("P(Draw)")
plt.title("VC Bookmaker Figure 4")
plt.legend(loc="lower center")
plt.show()
```

Although we plot the graphs with normalized odds, we removed bookmaker's margin, there is no significant bias for making money. While we decreasing bin intervals, actual outcome's volatility increases. We cannot specify exact odds difference in order to long run betting.
