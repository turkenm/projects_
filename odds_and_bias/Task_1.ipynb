Football matches can end in three different outcomes: the home team wins, the away team wins, and the match ends in a draw. Betting companies analyze the probabilities for these three different results according to many parameters such as the form of the teams, the form of the players, the statistics of the previous matches and determine a rate for each result. For example, let's say that a bookmaker sets the home, draw and away team odds as 1.30 - 4.70 - 6.80 respectively for the match result bet. In other words, for every 1 unit of money played, the betting company will pay the player 1.30 units if the home team wins the match, 4.70 units if it ends in a draw, and 6.80 units if the guest team wins. The probability of occurrence of events at these rates determined according to the results of the analysis can be found by using the formula p = 1 / Ratio. According to these determined rates, the probability of the home team to win is 1 / 1.30 = 0.769, the probability of the match to end in a draw is 1 / 4.70 = 0.213, the probability of the away team to win is 1 / 6.80 = 0.147. If we add these three possibilities together, we see that 0.769 + 0.213 + 0.147 = 1.129. According to probability theory, the sum of the probability of occurrence of all events in the sample space is 1. The excess of 0.129 here shows the profit margin of the bookmakers. That is, when betting on all three outcomes with equal payoffs, the player's winnings will be less than the total stake, and thus the bookmaker will make a profit regardless of the outcome of the match.

For this study, the data of the matches played in the English Premier League in the 2018-2019, 2019-2020, 2020-2021 seasons, provided by https://www.football-data.co.uk, were used. In the data set, there are the end-of-match statistics of the matches played in the relevant seasons and the odds for the match result bet determined by the 4 big betting companies in Europe. In order to start the analysis, first of all, the profit margin of the betting companies should be removed from the rates.
In an event with three outcomes;
pA, pB, pC are the probabilities of events A, B and C, respectively,
standardized probabilities.  
pA* = pA / (pA + pB + pC)  
pB* = pB / (pA + pB + pC)  
pC* = pC / (pA + pB + pC)  
pA* + pB* + pC* = 1  
available with formulas.
This process was also applied to 4 different betting companies in the data set.

# Poisson Distrubition 
We can use Scipy for calculate poisson probability, but for this time I want to write own function.

$$
P(λ,k) = \frac {e^-λ * λ ^k} {k!} 
$$

λ = sample mean  
k = number of goals  
If we multiply P and total number of games, we can find how many matches should finish with corresponding goal numbers.

```python
import math
```

```python
home_teams_lambda = hist_df["FTHG"].mean()
away_teams_lambda = hist_df["FTAG"].mean()
```

```python
def pmf (x, k):
   return(math.exp(-x) * x**k)/ math.factorial(k)*len(hist_df)
```

```python
poiss_home = []
poiss_away = []
for i in range(hist_df["FTHG"].max()):
    poiss_home.append(pmf(hist_df["FTHG"].mean(),i))
    poiss_away.append(pmf(hist_df["FTAG"].mean(),i))
```

```python
plt.hist(hist_df["FTHG"], bins = 20, label = "Hist", align = "mid")
plt.plot(poiss_home, color='red', marker='o', linestyle="dashed", label ="Matches")
plt.xticks(range(0,10,1))
plt.ylabel('Number of Games')
plt.xlabel('Home Goals')
plt.legend(loc="upper right");
```

```python
plt.hist(hist_df["FTAG"], bins=20, label = "Hist")
plt.plot(poiss_away, color='red', marker='o', linestyle="dashed", label ="Matches")
plt.xticks(range(0,10,1))
plt.ylabel('Number of Games')
plt.xlabel('Away Goals')
plt.legend(loc="upper right");
```
